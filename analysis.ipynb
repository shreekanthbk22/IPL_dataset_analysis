1.<b>Query to get the dismissal types with count.</b><br/>
'dismissal_kind' column contains type of dismissal and 'NA' if not dismissed.<br/>
As 'NA' is just indication of not dismissed and doesn't belong to the group of dismissal type, we are excluding 'NA'.

```
%sql
select distinct(`dismissal_kind`) as `dismissal_type`, count(*) as `no_of_dismissals` 
from `ipl_ball_by_ball` 
where `dismissal_kind`!="NA" 
group by `dismissal_kind`
```

2.<b>query to get list of ten fielders (fielder and count of catches) with the greatest
number of catches.</b><br/>
listed fielders with there number of catches in descending order and limited the rows to 10.

```
%sql
select distinct(`fielder`),count(*) as `no_of_catches` from `ipl_ball_by_ball` 
where `dismissal_kind`="caught" 
group by `fielder` 
order by `no_of_catches` desc 
limit 10
```

3.<b>listing of team and the number of sixes (6s) scored by each team.</b><br/>
As the sixes were not specified directly in dataset, we can infer them by ensuring non-boundary and extra runs were zero and batsman runs were 6 in a over.

```
%sql
select distinct(`batting_team`) as `Team`,count(*) as `no_of_sixes` 
from `ipl_ball_by_ball` 
where `batsman_runs`=6 and `non_boundary`=0 and `extra_runs`=0
group by `batting_team`
```

4.<b>query to find total runs scored by the batsman for each team he played. (Assume
all players are batsman)</b><br/>
selected columns of batsman and team he played and grouped by batsman name.<br/>
<br/>
As I observed team 'Rising Pune Supergiants' misspelled in batting team column.<br/>
And hence for batsman 'MS Dhoni' it shows 3 teams rather than 2.</h6>

```
%sql
select `batsman`, sum(`batsman_runs`) as `runs`, `batting_team` as `team_played` 
from `ipl_ball_by_ball` 
group by `batsman`,`batting_team` 
order by `batsman`
```

5.<b>query to get match_Id, Man of the match and runs scored by man of the match.</b><br/>
Retrieved all the runs scored by player of the match and summing up it we get total runs he scored.

```
%sql
select B.`match_id`, B.`player_of_match`, sum(A.`batsman_runs`) as `runs_scored` 
from `ipl_ball_by_ball` A 
inner join `ipl_matches` B 
on A.`match_id`=B.`match_id`
where A.`batsman`=B.`player_of_match`
group by B.`match_id`,B.`player_of_match`
```

6.<b>query to get list of team and boundary scored when the team won the toss and
decided to bat first. (boundary includes 4s and 6s)</b><br/>

```
%sql
select `batting_team` as `Team`, A.`batsman_runs` as boundaries
from `ipl_ball_by_ball` A
inner join `ipl_matches` B
on A.`match_id`=B.`match_id`
where `batsman_runs`=4 or `batsman_runs`=6 and `extra_runs`=0 and `non_boundary`=0 and B.`toss_winner`=A.`batting_team` and `toss_decision`="bat"
```

7.<b>query to return a report for Venue and City and average runs scored per match
in powerplay in 1st inning. [powerplay means 1 to 6th Over including 6th Over]</b><br/>
Here we joined three tables:&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<b>ipl_ball_by_ball</b>-->match_id<---<b>ipl_matches</b>--->venue_id<---<b>ipl_venue</b><br/>
* we retrieved total runs scored in each match in powerplay & 1st innings.
* grouped each match with respect to venue and took average of those runs.

```
%sql
select C.`venue_id`, C.`city`, avg(team_runs) as average_runs
from ipl_venue C
left join ipl_matches B
          on C.`venue_id`=B.`venue_id`
inner join
          (select `match_id`,sum(`total_runs`) as team_runs 
           from `ipl_ball_by_ball`
           where `inning`=1 and `overs`<=6
           group by `match_id`) A
           on B.`match_id`=A.`match_id`
group by C.`venue_id`, C.`city`
order by C.`venue_id`
```
